```python
import pandas as pd
from pathlib import Path

# Dossier contenant les fichiers DVF
dvf_dir = Path("C:/Users/AdrienSourdille/Desktop/Saint-Vincent-des-Landes-impact-study/data/raw/DVF")

# Liste tous les fichiers .parquet dans le dossier
files = list(dvf_dir.glob("*.parquet"))

# Charge et concatène tous les fichiers
df = pd.concat([pd.read_parquet(f) for f in files], ignore_index=True)

# Filtre le département 44
df_44 = df[df["code_departement"] == "44"]
```

```python
print(df_44["nature_mutation"].value_counts())

df_vente_44 = df_44[df_44["nature_mutation"] == "Vente"]
```

```python
print(df_vente_44["type_local"].value_counts())

df_vente_44_MA = df_vente_44[(df_vente_44["type_local"] == "Maison") | (df_vente_44["type_local"] == "Appartement")]
```

```python
print(df_vente_44_MA.info())
display(df_vente_44_MA)
```

```python
df_vente_44_MA["price_per_m2"] = df_vente_44_MA["valeur_fonciere"] / df_vente_44_MA["surface_reelle_bati"]
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Drop rows with missing coordinates
df_clean = df_vente_44_MA.dropna(subset=["longitude", "latitude"])

# Create geometry points
geometry = [Point(xy) for xy in zip(df_clean["longitude"], df_clean["latitude"])]

# Create GeoDataFrame
gdf = gpd.GeoDataFrame(df_clean, geometry=geometry, crs="EPSG:4326")  # WGS84
```

```python
display(gdf.head())
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
from geopy.distance import geodesic


# --- Define coordinates of the wind farm ---
windfarm_coords = (47.6670, -1.3170)  # (latitude, longitude)

# --- Calculate geodesic distance to wind farm (in km) ---
gdf["distance_km"] = gdf.apply(
    lambda row: geodesic((row["latitude"], row["longitude"]), windfarm_coords).km,
    axis=1
)

# --- Drop geometry to reduce size ---
df_final = gdf.drop(columns="geometry")
```

```python
import numpy as np

# df.distance_km is your computed distance
bins = [0, 1, 2, 5, 10, 20, 30, np.inf]
labels = ['0-1 km', '1-2 km', '2-5 km', '5-10 km', '10-20 km', '20-30 km', '>30 km']

df['distance_bucket'] = pd.cut(
    df['distance_km'],
    bins=bins,
    labels=labels,
    right=False
)
```

```python
# --- Save the result as lightweight CSV ---
output_path = "C:/Users/AdrienSourdille/Desktop/Saint-Vincent-des-Landes-impact-study/data/processed/dvf_vente_44_distances.csv"
df_final.to_csv(output_path, index=False)

print(f"✅ Saved file with distances to: {output_path}")
```
