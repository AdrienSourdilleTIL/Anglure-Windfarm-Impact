```python
import pandas as pd
from pathlib import Path

# Dossier contenant les fichiers DVF
dvf_dir = Path(r"C:\Users\AdrienSourdille\Desktop\Anglure-Windfarm-Impact\data\raw\DVF")

# Liste tous les fichiers .parquet dans le dossier
files = list(dvf_dir.glob("*.parquet"))

print(files)

# Charge et concatène tous les fichiers
df = pd.concat([pd.read_parquet(f) for f in files], ignore_index=True)

# Filtre le département 51
df_51 = df[df["code_departement"] == "51"]
```

```python
print(df_51["nature_mutation"].value_counts())

df_vente_51 = df_51[df_51["nature_mutation"] == "Vente"]
```

```python
print(df_vente_51["type_local"].value_counts())

df_vente_51_MA = df_vente_51[(df_vente_51["type_local"] == "Maison") | (df_vente_51["type_local"] == "Appartement")]
```

```python
print(df_vente_51_MA.info())
display(df_vente_51_MA)
```

```python
df_vente_51_MA["price_per_m2"] = df_vente_51_MA["valeur_fonciere"] / df_vente_51_MA["surface_reelle_bati"]
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Drop rows with missing coordinates
df_clean = df_vente_51_MA.dropna(subset=["longitude", "latitude"])

# Create geometry points
geometry = [Point(xy) for xy in zip(df_clean["longitude"], df_clean["latitude"])]

# Create GeoDataFrame
gdf = gpd.GeoDataFrame(df_clean, geometry=geometry, crs="EPSG:4326")  # WGS84
```

```python
display(gdf.head())
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
from geopy.distance import geodesic


# --- Define coordinates of the wind farm ---
windfarm_coords = (48.64249584971337, 3.787199273678923)  # (latitude, longitude)

# --- Calculate geodesic distance to wind farm (in km) ---
gdf["distance_km"] = gdf.apply(
    lambda row: geodesic((row["latitude"], row["longitude"]), windfarm_coords).km,
    axis=1
)

# --- Drop geometry to reduce size ---
df_final = gdf.drop(columns="geometry")
```

```python
display(df_final.head())
```

```python
import numpy as np
import pandas as pd  # in case it's not already imported

# Define distance buckets
bins = [0, 1, 2, 5, 10, 20, 30, np.inf]
labels = ['0-1 km', '1-2 km', '2-5 km', '5-10 km', '10-20 km', '20-30 km', '30-50 km']

# Create the bucketed column
df_final['distance_bucket'] = pd.cut(
    df_final['distance_km'],
    bins=bins,
    labels=labels,
    right=False
)

# Keep only transactions within 50 km of the wind farm
df_final = df_final[df_final["distance_km"] <= 50]
```

```python
# --- Save the result as lightweight CSV ---
output_path = "C:/Users/AdrienSourdille/Desktop/ANGLURE-WINDFARM-IMPACT/data/processed/dvf_vente_51_distances.csv"
df_final.to_csv(output_path, index=False)

print(f"✅ Saved file with distances to: {output_path}")
```
